```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from iedsp import util
from iedsp.evaluate import EvaluationManager
```

```python
scribe = EvaluationManager()
```

```python
history = util.load_from_pickle('exp/exp1.pickle')
history[1]['test'].keys()
```

```python
import numpy as np
returns = []
losses = []
turns = []
goals = []
epochs = list(history.keys())
for epoch in epochs:
    test_summary = history[epoch]['test']
    returns.append(np.mean(test_summary['return']))
    losses.append(np.mean(test_summary['loss']))
    turns.append(np.mean(test_summary['turn']))
    goals.append(np.mean(test_summary['goal']))
```

```python
max_test_summary = None
max_return_mean = -10000
for epoch in epochs:
    test_summary = history[epoch]['test']
    return_mean = np.mean(test_summary['return'])
    if return_mean > max_return_mean:
        max_return_mean = return_mean
        max_test_summary = test_summary
```

```python
scribe.pprint_summary(max_test_summary)
```

```python
fig, ax1 = plt.subplots()
ax1.set_title("DQN Policy Learning")
ax1.plot(epochs, returns, 'y-')
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Return(yellow)")

ax2 = ax1.twinx()
ax2.plot(epochs, goals, 'b-')
ax2.set_ylabel("Goal(blue)")
```
